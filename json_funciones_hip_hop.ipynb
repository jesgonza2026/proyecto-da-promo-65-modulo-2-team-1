```python
with open("hip_hop.json", "r",) as f: # Cargar el archivo JSON
    datos = json.load(f)
print("JSON cargado correctamente")
```

```python
# FUNCIÓN 1: Filtrar el JSON principal por rango de años (2018-2022)
def filtrar_rango_anios(json, inicio, fin): # Defino una función 
    rango_anios = [] # Creamos una lista vacía

    for elemento in json:  # Si el año buscado coincide lo agrega a la lista nueva
        if elemento["year"] in range(inicio, fin + 1):
            rango_anios.append(elemento)
    
    print("Se ha filtrado con éxito los años 2018-2022")

    return rango_anios # Saca el objetivo de la función 
```

```python
datos_18_22 = filtrar_rango_anios(datos, 2018, 2022)
```

```python
# FUNCIÓN 2: Obtener los géneros musicales
def filtrar_genero_musical(rango_anios):
    generos_musicales = set() # Usamos un set para que no haya duplicados

    for elemento in rango_anios:
        generos_musicales.add(elemento["genre"].lower()) # Para acceder a los datos independientemente de cómo estén escritos 

    print(f"Se han encontrado {len(generos_musicales)} géneros musicales:") # Mostrar la cantidad
    print("----")

    for elemento in generos_musicales:
        print(f"- {elemento}")
    
    return list(generos_musicales) # Lo sacamos en formato lista
```

```python
generos_musicales = filtrar_genero_musical(datos_18_22)
```

```python
# FUNCIÓN 3: Filtrar Hip-Hop por rango de años (2018-2022)

def filtrar_anios_genero(json, inicio, fin, genero): 
    archivos_hip_hop = []

    for elemento in json:  # Si el año y el género buscado está en el JSON, lo agrega a la lista creada
        if (elemento["year"] in range(inicio, fin + 1)) and (genero in elemento["genre"].lower()): # Si el genero buscado esta en el JSON, lo agrega a la lista creada
                archivos_hip_hop.append(elemento)
    
    print("Se ha filtrado con éxito")
    print("----")

    print(f"En 2018-2022 hay {len(archivos_hip_hop)} archivos de Hip-Hop")
    print("----")

    hip_hop_2018=[]
    hip_hop_2019=[]
    hip_hop_2020=[]
    hip_hop_2021=[]
    hip_hop_2022=[]

    for archivo in archivos_hip_hop:

        año_del_archivo = archivo["year"]

        if año_del_archivo == 2018:
                hip_hop_2018.append(archivo)

        elif año_del_archivo == 2019:
                hip_hop_2019.append(archivo)

        elif año_del_archivo == 2020:
                hip_hop_2020.append(archivo)

        elif año_del_archivo == 2021:
                hip_hop_2021.append(archivo)

        elif año_del_archivo == 2022:
                hip_hop_2022.append(archivo)

    print(f"En 2018 hay {len(hip_hop_2018)} archivos de Hip-Hop")
    print(f"En 2019 hay {len(hip_hop_2019)} archivos de Hip-Hop")
    print(f"En 2020 hay {len(hip_hop_2020)} archivos de Hip-Hop")
    print(f"En 2021 hay {len(hip_hop_2021)} archivos de Hip-Hop")
    print(f"En 2022 hay {len(hip_hop_2022)} archivos de Hip-Hop")
        
    return archivos_hip_hop
```

```python
lista_hip_hop = filtrar_anios_genero(datos, 2018, 2022, "hip-hop")
```

```python
# FUNCIÓN 4: Extraer infromación de Hip-Hop

def analizar_hip_hop(lista_hip_hop):
    datos_hip_hop = []
    artistas = set()

    for elemento in lista_hip_hop:
        informacion = { # Para extraer una información específica: ID, Nombre del artista, Nombre de la canción, Género musical y Año de lanzamiento
            "id" : elemento["id"],
            "artista" : elemento["artist_name"],
            "nombre" : elemento["track_name"],
            "genero" : elemento["genre"],
            "año" : elemento["year"]}
        datos_hip_hop.append(informacion)
        artistas.add(informacion["artista"])
    
    print("Se ha filtrado con éxito")
    print("----")
    
    print(f"Se han encontrado {len(artistas)} artistas diferentes en Hip-Hop") 
    
    return datos_hip_hop
```

```python
hip_hop = analizar_hip_hop(lista_hip_hop)
```

```python
df_hip_hop = pd.DataFrame(hip_hop) # Lo convertimos en una tabla
```
