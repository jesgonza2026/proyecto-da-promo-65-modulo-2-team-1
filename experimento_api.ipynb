```python
import requests

# 1. Lista limpia (sin listas anidadas)
nombre2 = ["Metallica", "Madonna"]
resultados = [] # Asegúrate de que esta lista esté definida antes

for nombre_artista in nombre2:
    print(f"Buscando información de: {nombre_artista}...")
    
    url = "http://ws.audioscrobbler.com/2.0/"
    
    # 2. Usamos 'nombre_artista' (el elemento actual del bucle)
    params = {
        "method": "artist.getInfo",
        "artist": nombre_artista,  
        "api_key": api_key_last_fm,
        "format": "json"
    }

    try:
        response = requests.get(url, params=params)
        
        # 3. Guardamos la respuesta JSON directamente
        if response.status_code == 200:
            data = response.json()
            resultados.append(data)
            print(f"Éxito: {nombre_artista} (Código: {response.status_code})")
        else:
            print(f"Algo falló: {nombre_artista} (Código: {response.status_code})")
        
    except Exception as e:
        print(f"Error de conexión con {nombre_artista}: {e}")

print(f"¡Proceso finalizado! Se han recuperado {len(resultados)} artistas.")
```
