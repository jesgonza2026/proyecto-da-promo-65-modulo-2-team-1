```python
with open('files/artistrock.json', 'r', encoding='utf-8') as f:
    lista_artistas = json.load(f)
```

```python
resultados = []
```

```python
for nombre_artista in lista_artistas:
    print(f"Buscando información de: {nombre_artista}...")
    url = "http://ws.audioscrobbler.com/2.0/"
    params = {
    "method": "artist.getInfo",
    "artist": nombre_artista,  # <--- Usamos la variable del bucle, no la lista completa
    "username": usuario,
    "format": "json"
    }

    try:
        # La petición ocurre DENTRO del bucle
        response = requests.get(url, params=params)
        data = response.json()
        
        # Guardamos el resultado en nuestra lista
        resultados.append(data)
        
    except Exception as e:
        print(f"Error con {nombre_artista}: {e}")
    
    # -----------------------------------------------

# Al finalizar (fuera del bucle), imprimimos el total
print(f"¡Proceso finalizado! Se han recuperado {len(resultados)} artistas.")
```
