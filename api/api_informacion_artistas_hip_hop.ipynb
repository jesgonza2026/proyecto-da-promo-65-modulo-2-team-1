```python
# FUNCIÓN 1: Extraer los Artistas del JSON de Hip-Hop

def artistas_hip_hop (lista_hip_hop):
    lista_artistas = []

    print(f"Iniciando comparación de datos para {len(lista_hip_hop)} artistas")

    for elemento in lista_hip_hop:
        informacion = {"artista" : elemento["artist_name"]} # Para extraer una información específica: Nombre del artista, 
        lista_artistas.append(informacion)
    
    print("----")
    print("Se ha filtrado con éxito")
    
    return lista_artistas
```

```python
# FUNCIÓN 2: Extraer información de la API teniendo en cuenta los artistas de Hip-Hop (En mi caso, no lo hice con el último JSON extraído y guardado)

def filtrar_artistas_comunes (lista_artistas_hip_hop):
    artistas_comun = []
  
    print(f"Iniciando comparación de datos para {len(lista_artistas_hip_hop)} artistas")
    print("----")

    for elemento in lista_artistas_hip_hop:
        print(f"Buscando información de: {elemento}")
        url = "http://ws.audioscrobbler.com/2.0/" # Url 
        params = { # Para configurar el diccionario 
            "method": "artist.getInfo", # Para mirar dentro de esa lista específica del URL
            "artist": elemento,  # Usamos la variable del bucle, no la lista completa para buscar el nombre del artista
            "api_key" : "e313d3e1e9201b89cf4199ad1591d3d7", # Identificador de 'Last.fm'
            "format": "json"} # Para que la respuesta sea procesable por Python  
            
        try:
            
            response = requests.get(url, params=params) # La petición ocurre DENTRO del bucle
            datos_api = response.json() 
            print(f"Se ha logrado la conexion: {response.status_code}")
            print("----")

            if "artist" in datos_api:
                artistas_comun.append(elemento)  # Guardamos el resultado en nuestra lista
            
        except Exception as e:
            print(f"Error con {elemento}: {e}")
            print("----")

    print(f"¡Proceso finalizado! Se han recuperado {len(artistas_comun)} artistas en común entre el JSON y en la API")

    return artistas_comun
```

```python
# FUNCIÓN 3: Extraer información de los Artistas

def analizar_artistas (lista_artistas):
    datos_artistas = []
    lista_errores = [] # Lista para los fallos

    print(f"Iniciando comparación de datos para {len(lista_artistas_hip_hop)} artistas")
    print("----")

    for elemento in lista_artistas:
        nombre_artista = elemento["artista"] # Para iterar directamente por la Key del diccionario 

        url = "http://ws.audioscrobbler.com/2.0/" 
        params = { 
            "method": "artist.getInfo",
            "artist": nombre_artista, 
            "api_key" : "e313d3e1e9201b89cf4199ad1591d3d7",
            "format": "json"}
            
        try:
            response = requests.get(url, params=params) 
            datos_api = response.json() 
            print(f"Se ha logrado la conexion: {response.status_code}")
            print("----")

            if "artist" not in datos_api: # Si la API responde pero el artista no existe
                lista_errores.append({"artista" : nombre_artista})
                continue

            try: # Si el artista existe, intentamos sacar sus datos
                artistas_similares = datos_api["artist"]["similar"]["artist"] # Para iterar directamente desde las Key del diccionario 
                nombres_similares = [art["name"] for art in artistas_similares[:5]] # Para encontrar más de 1 artista similar, hasta 5
                
                informacion = { 
                        "artista": datos_api["artist"]["name"],
                        "audiencia": datos_api["artist"]["stats"]["listeners"],
                        "reproducciones" : datos_api["artist"]["stats"]["playcount"],
                        "similares" : nombres_similares, 
                        "biografía" : datos_api["artist"]["bio"]["summary"]}
                datos_artistas.append(informacion)

            except Exception as e: # Exception = "Captura cualquier error que ocurra"
                lista_errores.append({"artista" : nombre_artista})
                    
        except Exception as e:
            print(f"Error con {elemento}: {e}")
            print("----")

    print(f"Se han encontrado {len(lista_errores)} artistas que han dado algún error")
    print("----")
    print(f"Se han filtrado con éxito {len(datos_artistas)} artistas")

    return datos_artistas
```

```python
artistas = analizar_artistas(lista_artistas)
```

```python
df_artistas = pd.DataFrame(artistas)
```
