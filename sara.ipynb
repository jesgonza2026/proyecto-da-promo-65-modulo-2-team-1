```python
# sacar de fuente.json todos los archivos que lleva 'indie'
archivos_indie = [] # en esta variable introduzco todos los archivos que son de genre=indie

for elemento in fuente_json:
    
    if 'indie' in elemento['genre'].lower():
        archivos_indie.append(elemento)

print(archivos_indie)
```

```python
archivos_indie
```

```python
#creo variables para pasar cada archivo por años
indie_2018=[]
indie_2019=[]
indie_2020=[]
indie_2021=[]
indie_2022=[]

# lo meto por un bucle for para iterar por el 

for archivo in archivos_indie:

    año_del_archivo = archivo['year']

    if año_del_archivo == 2018:
            indie_2018.append(archivo)

    elif año_del_archivo == 2019:
            indie_2019.append(archivo)

    elif año_del_archivo == 2020:
            indie_2020.append(archivo)

    elif año_del_archivo == 2021:
            indie_2021.append(archivo)

    elif año_del_archivo == 2022:
            indie_2022.append(archivo)

# Ahora puedes ver cuántos hay en cada una

print(f"En 2018 hay {len(indie_2018)} archivos indie.")
 
print(f"En 2019 hay {len(indie_2019)} archivos indie.")
print(f"En 2020 hay {len(indie_2020)} archivos indie.")
print(f"En 2021 hay {len(indie_2021)} archivos indie.")
print(f"En 2022 hay {len(indie_2022)} archivos indie.")


```
