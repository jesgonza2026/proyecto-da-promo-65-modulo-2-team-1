```python
import json # Importar la librería para poder leer documentos en formato JSON

with open('file1.json') as f1, open('file2.json') as f2: # Para abrir y leer TODOS los archivos que queremos unir en un solo JSON
    data1 = json.load(f1) # Creamos una variable por cada JSON
    data2 = json.load(f2)

combined_data = data1 + data2 # Crear otra variable para combinar todos los JSON

with open('combined.json', 'w') as f: # Guardar y extraer la variable con TODOS los JSON en formato JSON
    json.dump(combined_data, f)
```

```python
# --- EJEMPLO ---
# Utilizo la forma hecha en el meeting con las chicas del PROYECTO - Mod2

# PASO 1: Abrimos librerias necesarias
import json
import pandas as pd

# PASO 2: Abrimos 1x1 TODOS los archivos JSON 'mini_json_genero'
with open("mini_json_hip_hop.json", "r",) as f: # Cargar el archivo JSON de Hip-Hop
    datos_hip_hop = json.load(f)

with open("mini_json_rock.json", "r",) as f: # Cargar el archivo JSON de Rock
    datos_rock = json.load(f)

with open("mini_json_latin.json", "r",) as f: # Cargar el archivo JSON de Latin
    datos_latin = json.load(f)

with open("mini_json_pop.json", "r",) as f: # Cargar el archivo JSON de Pop
    datos_pop = json.load(f)
    
with open("mini_json_rap.json", "r",) as f: # Cargar el archivo JSON de Rap
    datos_rap = json.load(f)

# PASO 3: Unimos TODOS los MiniJSON
datos_combinados =  datos_hip_hop + datos_rock + datos_latin + datos_pop + datos_rap 

# PASO 4: Creamos un DataFrame para visualizar el contenido
df_datos = pd.DataFrame(datos_combinados)

df_datos

# PASO 5: Guardamos y extraemos el JSON General en formato JSON
with open("json_general_artistas.json", "w") as f: 
    json.dump(datos_combinados, f, indent = 4, ensure_ascii =  False)
```
