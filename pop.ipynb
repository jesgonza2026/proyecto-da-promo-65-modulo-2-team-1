Hemos visto en un dataframe como es la info que tenemos en el definitivo.json.Ahora queremos filtrar genero y años.
Para ello quiero usar un bucle for que itere por cada uno de los diccionarios y añada a una lista la información.
Para ello voy abrir el fichero definitivo.json con Python en lugar de con PANDAS.Pa ra ver la info ,lo veo más claro en un dataFrame (PANDAS),sin emargo para trabajarlo prefiero formato JSON por mi conocimiento hasta el momento(trabajar con listas y diccionarios).

```python
import json

#Abrimos el definitivo.json con python
with open("Files/definitivo.json") as file:
    datos_json_definitivo=json.load(file) #Ahora datos_json_definitivo es una variable en Python con todos los datos del definitivo.json.Cada elemento es un diccionario que representa una canción
```

```python
type(datos_json_definitivo)#Comprobamos con qué voy a tener que trabajar, que es lo que tengo.
```

```python
(datos_json_definitivo)[0]#Accedemos al primer elemento de la lista para ver que formato tiene, 
```

```python
#Filtramos usando un bucle for y if

minijson_pop = [] #Creamos una lista vacía para guardar canciones pop 2018-2022

for cancion in datos_json_definitivo:
    # Convertimos el género a minúsculas para evitar mayúsculas
    genero = cancion["genre"].lower()
    
    # Revisamos si cumple los filtros
    if genero == "pop" and 2018 <= cancion["year"] <= 2022:
        # Agregamos solo las columnas que queremos
        minijson_pop.append({
            "id": cancion["id"],
            "artist_name": cancion["artist_name"],
            "track_name": cancion["track_name"],
            "genre": cancion["genre"],
            "year": cancion["year"]
        })
        #break --------------------------#le pongo un break para que me muestre una cancion /un elemento y ver que está correcto
        # y se lo quito porque he comprobado que me da lo que quiero y quiero que se ejecute el bucle y añadala todas las canciones a la lista minijson_pop
```

```python
# Guardar minijson_pop en un archivo llamado popminijson.json

with open("Files/popminijsonFINAL.json", "w", encoding="utf-8") as f: ##encoding="utf-8" → asegura que los caracteres especiales (acentos, ñ, emojis, etc.) se guarden correctamente.PARA ABRIR ARCHIVO
    json.dump(minijson_pop, f, indent=4, ensure_ascii=False) #ensure_ascii=False → asegura que los caracteres especiales (acentos, ñ, emojis, etc.) se guarden correctamente.PARA CERRAR
```
