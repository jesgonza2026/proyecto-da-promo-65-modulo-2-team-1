```python
# sacar de fuente.json todos los archivos que lleva 'pop'
archivos_pop = [] # en esta variable introduzco todos los archivos que son de genre=pop

for elemento in fuente_json:
    
    if 'pop' in elemento['genre'].lower():
        archivos_pop.append(elemento)

print(archivos_pop)
```

```python
archivos_pop
```

```python
#creo variables para pasar cada archivo por años
pop_2018=[]
pop_2019=[]
pop_2020=[]
pop_2021=[]
pop_2022=[]

# lo meto por un bucle for para iterar por el 

for archivo in archivos_pop:

    año_del_archivo = archivo['year']

    if año_del_archivo == 2018:
            pop_2018.append(archivo)

    elif año_del_archivo == 2019:
            pop_2019.append(archivo)

    elif año_del_archivo == 2020:
            pop_2020.append(archivo)

    elif año_del_archivo == 2021:
            pop_2021.append(archivo)

    elif año_del_archivo == 2022:
            pop_2022.append(archivo)

# Ahora puedes ver cuántos hay en cada una

print(f"En 2018 hay {len(pop_2018)} archivos pop.")
 
print(f"En 2019 hay {len(pop_2019)} archivos pop.")
print(f"En 2020 hay {len(pop_2020)} archivos pop.")
print(f"En 2021 hay {len(pop_2021)} archivos pop.")
print(f"En 2022 hay {len(pop_2022)} archivos pop.")


```

```python
with open("Files/archivo.json") as archivo_json:
    fuente_json = json.load(archivo_json)
```

```python
artistas_unicos = sorted({ #Creamos un set (colección de elementos únicos, sin duplicados) y le ponemos un sorted porque los sets no tienen orden.
    elemento["artist_name"]
    for elemento in fuente_json #Recorre cada elemento de nuestro JSON (cada canción).
    if 2018 <= elemento["year"] <= 2022 and elemento["genre"].lower() == "pop" # Filtra solo las canciones cuyo año esté entre 2018 y 2022.Nos aseguramos de que asegura que solo se consideren elementos(canciones) de pop, sin importar mayúsculas/minúsculas
})

print("Artistas únicos de pop entre 2018 y 2022:")
for artista in artistas_unicos:
    print(artista)
```

```python
# Guardar en un archivo JSON - #Guardamos minijson_pop


with open("artistas_pop_2018_2022.json", "w", encoding="utf-8") as minijson_pop: #utf-8 asegura que los caracteres especiales (como acentos) se guarden correctamente
    json.dump(artistas_unicos, minijson_pop, ensure_ascii=False, indent=4) #ensure_ascii=False permite que el JSON guarde los caracteres especiales correctamente en su forma natural(tildes,ñ etc)indent :Cada elemento está en su propia línea y sangrado con 4 espacios
```
