```python
ruta_correcta = "Files/definitivo.json" 

with open(ruta_correcta, "r", encoding='utf-8') as archivo:
    datos = json.load(archivo)
```

```python
datos
```

```python
#vamos a normalizar a ver que tal
df = pd.json_normalize(datos)
```

```python
df
```

```python
#es muy grande entonces estamos viendo el size
dimensiones = df.shape
print(dimensiones)
```

Sacamos los registros con los anos usando between

```python
df_rango = df[df['year'].between(2018, 2022)]

print(f"He encontrado {df_rango.shape[0]} registros en ese periodo.")
print(df_rango.head())
```

```python
#Filtramos y luego contamos géneros
conteo_generos_20s = df[df['year'].between(2018, 2022)]['genre'].value_counts()

print("Géneros más populares entre 2018 y 2022:")
print(conteo_generos_20s)
```

Reciclo el codigo de sara

```python
df_rock = df_rango[df_rango['genre'].str.contains('rock', case=False, na=False)].copy()
```

```python
df_rock
```

```python
conteo_por_ano = df_rock['year'].value_counts().sort_index()

print("Conteo de archivos Rock por año:")
print(conteo_por_ano)
```
