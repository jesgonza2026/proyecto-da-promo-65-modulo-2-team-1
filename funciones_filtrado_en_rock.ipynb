```python
import json
import pandas as pd

# Funciones
def get_genero_data(json, genero):
    archivos_genero = [] # defino funcion para encontrar genero

    for elemento in json:
        if genero in elemento['genre'].lower(): #si el genero esta en la lista, lo agrega
            archivos_genero.append(elemento)
    
    return archivos_genero

def filtrar_anios(json, inicio, fin): 
    cumplen = []

    for elemento in json:  #creo funcion para definir los años con un range+1 y lo agrega a la lista
        if elemento['year'] in range(inicio, fin + 1):
            cumplen.append(elemento)

    return cumplen

# Abro json
with open("Files/definitivo.json", "r") as fichero: # introducimos el nombre de nuestro archivo json y "r" porque lo vamos a leer(read)
    canciones = json.load(fichero)
```

```python

# Uso funciones para filtrar
latin_genre = get_genero_data(canciones, 'latin')

print("Las cantidad de canciones latinas es: ", len(latin_genre)) #podemos cambiar las variables de genero y rango de años en la misma funcion.

latin_fitradas = filtrar_anios(latin_genre, 2018, 2022)

# Muestro datos con panda
pd.DataFrame(latin_fitradas)
```

```python
# Uso funciones para filtrar
rock_genre = get_genero_data(canciones, 'rock')

print("Las cantidad de canciones de rock es: ", len(rock_genre)) #podemos cambiar las variables de genero y rango de años en la misma funcion.

rock_fitradas = filtrar_anios(rock_genre, 2018, 2022)

# Muestro datos con panda
pd.DataFrame(rock_fitradas)
```
