```python
#me ha tocado rap
#tengo que separar por año y genero,  alla vamos !!!
rap= []
for  elemento in dato1:
    
    if 'rap' in elemento['genre'].lower():
        rap.append(elemento)
```

```python
rap
#aqui tengo todas las  canciones de rap del documento
   
```

```python
#creo variables para pasar cada archivo por años
rap_2018=[]
rap_2019=[]
rap_2020=[]
rap_2021=[]
rap_2022=[]

# lo meto por un bucle for para iterar por el 

for archivo in rap:

    año_del_archivo = archivo['year']

    if año_del_archivo == 2018:
           rap_2018.append(archivo)

    elif año_del_archivo == 2019:
            rap_2019.append(archivo)

    elif año_del_archivo == 2020:
            rap_2020.append(archivo)

    elif año_del_archivo == 2021:
           rap_2021.append(archivo)

    elif año_del_archivo == 2022:
            rap_2022.append(archivo)

# Ahora puedes ver cuántos hay en cada una

print(f"En 2018 hay {len(rap_2018)} archivos rap.")
print('---------------------------------------------------------------')
print(f"En 2019 hay {len(rap_2019)} archivos rap.")
print('---------------------------------------------------------------')
print(f"En 2020 hay {len(rap_2020)} archivos rap.")
print('---------------------------------------------------------------')
print(f"En 2021 hay {len(rap_2021)} archivos rap.")
print('---------------------------------------------------------------')
print(f"En 2022 hay {len(rap_2022)} archivos rap.")
print('---------------------------------------------------------------')
```
