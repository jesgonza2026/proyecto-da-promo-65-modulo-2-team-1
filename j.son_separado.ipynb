```python

import pandas as pd
import json
```

```python
with open("Files/definitivo.json", "r") as fichero: # traigo el fichero
    datos_personaje = json.load(fichero)
```

```python
#declaro el dicionario donde lo voy a meter

dicionario_personajes ={'id':[],
                        'nombre_artista':[],
                        'nombre_cancion':[],
                        'año_lanzamiento':[],
                        'genero_musical':[]}
```

```python
dicionario_personajes['id'].append(primer_dato.get('id'))
dicionario_personajes['nombre_artista'].append(primer_dato.get('artist_name'))
dicionario_personajes['nombre_cancion'].append(primer_dato.get('track_name'))
dicionario_personajes['año_lanzamiento'].append(primer_dato.get('year'))
dicionario_personajes['genero_musical'].append(primer_dato.get('genre'))

print(dicionario_personajes)
df_de_1= pd.DataFrame(dicionario_personajes)

df_de_1
```

```python
# funcion 

def almacenar_datos(datos_personaje, dicionario_personajes):
    try:
        for dato in datos_personaje:
            
            dicionario_personajes['id'].append(dato.get('id'))
            dicionario_personajes['nombre_artista'].append(dato.get('artist_name'))
            dicionario_personajes['nombre_cancion'].append(dato.get('track_name'))
            dicionario_personajes['año_lanzamiento'].append(dato.get('year'))
            dicionario_personajes['genero_musical'].append(dato.get('genre'))
        
        return pd.DataFrame(dicionario_personajes)

    except Exception as e:
        print(f"Error en la lista: {e}")
        return None

df_final= almacenar_datos(datos_personaje, dicionario_personajes)

```
