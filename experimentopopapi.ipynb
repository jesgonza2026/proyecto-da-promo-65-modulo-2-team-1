```python
import requests
import time

# 1. Tu configuración (Revisa bien tu API Key)
api_key_last_fm = "dffc96fb5e37b49d6382379a54a66094"
headers = {'User-Agent': 'MiBootcampApp/1.0'} # <--- Esto es vital para evitar el 403
url = "http://ws.audioscrobbler.com/2.0/"

diccionario_pop = {"artista" : [], "audiencia" : [], "reproducciones" : [], "similares" : [], "biografia": []}

for artista in lista_artistas_pop[0:3]:
    # 2. NO pongas los datos en la URL manualmente. 
    # Deja que 'requests' se encargue de limpiar los espacios y tildes.
    parametros = {
        "method": "artist.getinfo",
        "artist": artista,
        "api_key": api_key_last_fm,
        "format": "json"
    }

    try:
        # Pasamos los parámetros por separado
        response = requests.get(url, params=parametros, headers=headers)
        
        if response.status_code == 200:
            datos_api = response.json()
            
            if 'artist' in datos_api:
                info = datos_api['artist']
                
                diccionario_pop["artista"].append(info.get('name', artista))
                diccionario_pop["audiencia"].append(info['stats'].get('listeners', 0))
                diccionario_pop["reproducciones"].append(info['stats'].get('playcount', 0))
                
                # Extraemos similares (si existen)
                similares = [s['name'] for s in info.get('similar', {}).get('artist', [])]
                diccionario_pop["similares"].append(similares)
                
                # Biografía
                bio = info.get('bio', {}).get('content', 'Sin biografía')
                diccionario_pop["biografia"].append(bio)
                
                print(f"✅ {artista} recuperado correctamente.")
            else:
                print(f"⚠️ {artista} no existe en Last.fm.")
        
        elif response.status_code == 403:
            print(f"❌ Error 403 en {artista}: Revisa tu API Key o el User-Agent.")
            break # Si es 403, mejor paramos para investigar
            
        # Un pequeño respiro para no saturar la API
        time.sleep(0.1)

    except Exception as e:
        print(f"💥 Error inesperado con {artista}: {e}")

print("Proceso terminado.")
```
